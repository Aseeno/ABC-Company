```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import PowerTransformer
```

```python
# Display initial rows and check for missing values
print("Initial Dataset Overview:")
print(df.info())
print("\nMissing Values Count:")
print(df.isnull().sum())
```

```python
# Step 1: Handle Missing Values
# Using mean imputation for numerical columns and most frequent imputation for categorical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Imputers
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

# Applying imputers
df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

```python
# Step 2: Check and Remove Outliers
# Using the IQR (Interquartile Range) method
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Removing outliers for all numerical columns
for col in num_cols:
    original_count = len(df)
    df = remove_outliers(df, col)
    print(f"{col}: Removed {original_count - len(df)} outliers.")
```

```python
# Step 3: Address Skewed Data in Numerical Features
# Checking skewness
print("\nSkewness of numerical features:")
skewness = df[num_cols].skew()
print(skewness)

# Applying power transformation for highly skewed features
transformer = PowerTransformer(method='yeo-johnson')  # Handles both positive and negative values
skewed_cols = skewness[abs(skewness) > 1].index  # Features with skewness > 1

if len(skewed_cols) > 0:
    df[skewed_cols] = transformer.fit_transform(df[skewed_cols])
    print(f"\nApplied Power Transformation to skewed columns: {list(skewed_cols)}")
else:
    print("\nNo significantly skewed features detected.")

```

```python
# Step 4: Visualizing the Cleaned Data
# Checking the distribution of a sample numerical column
sample_col = num_cols[0]  # Replace with a specific column if needed
sns.histplot(df[sample_col], kde=True)
plt.title(f"Distribution of {sample_col} After Cleaning")
plt.show()

# Final overview of the cleaned dataset
print("\nCleaned Dataset Overview:")
print(df.info())
```

```python
from scipy.stats import boxcox
import numpy as np

# Function to apply Box-Cox Transformation (only for positive values)
def apply_boxcox(column):
    df[column], _ = boxcox(df[column] + 1)  # Adding 1 to handle zero values

# Re-check skewness after initial Power Transformation
print("\nRe-checking Skewness of Numerical Features:")
skewness = df[num_cols].skew()
print(skewness)
```

```python
# Identify highly skewed columns again
highly_skewed = skewness[abs(skewness) > 1].index
print(f"\nColumns still highly skewed after Power Transformation: {list(highly_skewed)}")
```

```python
# Apply Box-Cox Transformation to remaining highly skewed columns
for col in highly_skewed:
    if (df[col] > 0).all():  # Box-Cox works for strictly positive values
        apply_boxcox(col)
        print(f"Applied Box-Cox Transformation to column: {col}")
    else:
        # Use log transformation for non-positive columns
        df[col] = np.log1p(df[col] - df[col].min() + 1)
        print(f"Applied Log Transformation to column: {col}")

# Final Skewness Check
final_skewness = df[num_cols].skew()
print("\nFinal Skewness of Numerical Features:")
print(final_skewness)
```
