```python
# Count the number of employees in each team
team_counts = df['Team'].value_counts()

# Calculate the percentage split
total_employees = len(df)
team_percentages = (team_counts / total_employees) * 100

# Create a DataFrame for better handling
team_distribution = pd.DataFrame({
    'Number of Employees': team_counts,
    'Percentage': team_percentages
})

print(team_distribution)
```

```python
import matplotlib.pyplot as plt

# Bar plot for number of employees
plt.figure(figsize=(10, 6))
team_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Employees in Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Pie chart for percentage split
plt.figure(figsize=(10, 8))
team_percentages.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired(range(len(team_percentages))))
plt.title('Percentage Split of Employees by Team')
plt.ylabel('')  # Hide y-label
plt.tight_layout()
plt.show()
```

```python
# Both the pie and bar graph representation reveals the total workforce in each teams. In the above given charts reveals team A is largest by gaining 40%of total workforce.Then team Band C follow with 25 and 20 %.Both the chart illusrates percentage distribution and help the readers to collect data easily.
```

```python
print(df['Age'].describe())
```

```python
# Define age bins and labels
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ['<25', '25-34', '35-44', '45-54', '55-64', '65+']

# Create age groups
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_counts = df['age_group'].value_counts()

# Calculate the percentage split
total_employees = len(df)
age_group_percentages = (age_group_counts / total_employees) * 100

# Create a DataFrame for better handling
age_distribution = pd.DataFrame({
    'Number of Employees': age_group_counts,
    'Percentage': age_group_percentages
})

print(age_distribution)
```

```python
import matplotlib.pyplot as plt

# Bar plot for number of employees by age group
plt.figure(figsize=(12, 8))
age_group_counts.sort_index().plot(kind='bar', color='coral')
plt.title('Number of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Pie chart for percentage distribution by age group
plt.figure(figsize=(12, 10))
age_group_percentages.sort_index().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired(range(len(age_group_percentages))))
plt.title('Percentage Distribution of Employees by Age Group')
plt.ylabel('')  # Hide y-label
plt.tight_layout()
plt.show()
```

```python
#The givn data shows the most predominant age group amongemployees is 25-34 years old by comprising 40% of workforce.This heips the readers to understand workforce demographics.
```

```python
# Calculate total salary expenditure by team
team_salary_expenditure = df.groupby('Team')['Salary'].sum()

# Calculate total salary expenditure by position
position_salary_expenditure = df.groupby('Position')['Salary'].sum()

# Create DataFrames for better handling
team_salary_df = pd.DataFrame({
    'Total Salary Expenditure': team_salary_expenditure
})

position_salary_df = pd.DataFrame({
    'Total Salary Expenditure': position_salary_expenditure
})

print("Team Salary Expenditure:\n", team_salary_df)
print("Position Salary Expenditure:\n", position_salary_df)
```

```python
import matplotlib.pyplot as plt

# Bar plot for salary expenditure by team
plt.figure(figsize=(12, 8))
team_salary_expenditure.sort_values().plot(kind='bar', color='royalblue')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
# Bar plot for salary expenditure by position
plt.figure(figsize=(12, 8))
position_salary_expenditure.sort_values().plot(kind='bar', color='seagreen')
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
 #In theese above chart reveals Team A has the hieghest total salary expenditure and Senior manager role has heighest expenditure.
```

```python
import matplotlib.pyplot as plt

# Scatter plot for age vs salary
plt.figure(figsize=(12, 8))
plt.scatter(df['Age'], df['Salary'], alpha=0.5, c='blue', edgecolors='w', s=100)
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Scatter plot with regression line
plt.figure(figsize=(12, 8))
sns.regplot(x='Age', y='Salary', data=df, scatter_kws={'s':100, 'alpha':0.5}, line_kws={'color':'red'})
plt.title('Age vs Salary with Regression Line')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#In theese we analysis the correlation between age, salary and employees.Here the scatter plot shows that while there is a general upward trend in salary with agethere is considerable variation.
```
